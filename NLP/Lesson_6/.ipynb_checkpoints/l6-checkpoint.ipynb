# CNN

1.关于神经网络中正则化：每个参数的影响降低   
L1: $ L = loss + \sum{|W|} $  可能减少参数的数量，一些参数的值为0  
L2: $ L = loss + \sum{W^2} $  参数的值会非常小，但不会变0  
2.dropout：随机抹掉一些神经元  
3.拟合一个数据的分布  
4.BN:为了防止数据偏差，做一个归一化，再加两个参数  
5.卷积的好处，参数共享，平移不变性，检测边缘  
6.padding 卷积之前 在矩阵外部加一圈像素值 为了保持维度 并且增强边缘像素点的应用  
7.假设padding p padding维度 n矩阵维度 f核维度  最后得到的矩阵维度(n + 2p - f + 1)  
8.Valid Padding 正常卷积，没有padding  最后矩阵 n - f + 1  
9.Same Padding 维度保持不变，加了padding   p = (f - 1) / 2  
10.步长strides 最后矩阵维度  [(n + 2p - f) / s + 1]  
11.感受视野：用来计算输出值的像素值范围。 随着卷积层数的增加，每个像素的感受视野会越来越大，最后可能1个像素点代表整张图  
12.pooling：最大值和平均值，好处是降维

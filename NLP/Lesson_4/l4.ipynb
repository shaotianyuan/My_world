### Cost Functions

1，MSE  预测值与真实值的差的平方的和的平均（回归问题） $$ L = -1/m \sum_{n=1}^{m}{(y - \hat{y}) ^ 2} $$

2，绝对值代替平方（回归问题）$$ L = -1/m \sum_{n=1}^{m}{|y - \hat{y}|}$$

3，二分类 $$ L = -1/m [ y\sum_{n=1}^{m}{log(\hat{y})}+(1-y)\sum_{n=1}^{m}{log(1 - \hat{y})} ]$$

4，交叉熵 & softmax （多分类）  $$ softmax = e ^ a / \sum_{n=1}^{m}{e ^ a}$$ 
$$ crossentropy = -\sum_{n=1}^{m}{ylog(s_j)} $$ 

#### 举例：  
y（onehot的形式 ） = [0, 0, 1, 0]  
sj= softmax的值
$$L(交叉熵) = -( 0* logS_1+ 0*logS_2 + 1*logS_3 + 0*logS_4) = -logS_3 = -log (e^n/E) $$

#### cross entropy是更通用的说法 而逻辑回归的loss function 是二分类的其中一种情况

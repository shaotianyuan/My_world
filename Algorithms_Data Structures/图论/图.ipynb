## 图论：

1，图对比树，是更为一般的数据结构，更具有抽象性  
2，图由顶点、边、边的权重组成  
3，路径：由边依次连接顶点的序列  
    1）无权路径长度为边的数量   
    2）带权路径的长度为所有边的权重和  
4，圈：首位顶点相同的路径  
    1）DAG：有向无圈图






## API

Graph():创建一个空的图  
addVertex(vert):将顶点vert加入图中  
addEdge(fromVert, toVert, weight):添加有向边(可以带有权重)   
getVertex(vkey):查找名称为vkey的顶点   
getVertices():返回图中所有顶点的列表   
in:顶点是否存在，返回True/False

## ADT Graph

1，主要有两种实现方式，邻接矩阵和邻接表  

### 邻接矩阵表示：
1，顶点数量n， n * n 的矩阵  
2，边表示为，矩阵中的行列值   
3，边数量非常少的时候效率低下






### 邻接列表
1，每个顶点一个id， 相连的边存在一个字典里



```python
# 顶点的类
class Vertex: 
    def __init__(self, key):
        self.id = key
        self.connectedTo = {}
        
    def addNeighbor(self, nbr, weight=0):
        #nbr是指向顶点的key
        self.connectedTo[nbr] = weight
        
    def __str__(self):
        return str(self.id) + 'connectedTo:' + str([x.id for x in self.connectedTo])
    
    def getConnections(self):
        return self.connectedTo.keys()
    
    def getId(self):
        return self.id
    
    def getWeight(self, nbr):
        return self.connectedTo[nbr]
```

## 初始化策略


1. 非对称（对称位置，反传梯度一样，更新值一样，所以不可以初始值一样，否则都相同没意义）
2. 不能全0、全1
3. 不能太大（会有梯度消失现象）
4. 不能太小（梯度同样会消失，relu参数消失）
5. Xavier：NN的每组输入组与输出组的方差相近
6. kaiming（MSRA）：针对relu激活，在Xavier上做了优化的初始化方式



## 图片预处理

1. 传统数据初始化方法：减去平均值，除以标准差
2. PCA：主成分分析


## 优化器 optimization

1. SGD 梯度下降（核心）
2. 全部数据：内层不够、局部loss极小值
3. mini-batch，每次取一部分数据
4. Momentum：冲量，给到一定的惯性，上一次的theta * garmma
5. Nesterov：用超前的梯度，更新冲量（基于动量）
6. Adagrad：学习率自适应，把历史梯度考虑进来，学习率除以历史梯度的平方
7. RMSProp：对Ada改进（S是全部历史梯度平方和），加权与更新时间有关
8. ADAM：自适应+冲量

总结：
1. Adam 收敛比较快
2. 最后阶段用SGDM效果比较好

## Evaluation

1. 分类检测指标：  
Accuracy：不同类别，地位相同(准确：对了多少个)  
Precision（精准）/recall：不同类别，地位不同，关键类别的准确率与召回率

2. 混淆矩阵：  
ACC(准确率): TP + TN / TP + FP + FN + TN  
RECALL(召回率): TP / TP + FN  
PRECISION(精准率): TP / TP + FP  


ROC: 受试者，操作特性曲线
AVC: 曲线下的面积


### 学习率调整
1. adam、adagrad 自适应 学习率调制不是必须的
2. SGDM 最好加学习率调整
3. 余弦退火（周期性）：图形余弦的一半，为了突破局部极小值点
4. 查看loss更新情况：很多轮没有变化，lr变小

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
def my_show(img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
img1 = cv2.imread('test3.png')
img2 = cv2.imread('test1.jpeg')
```

```python
sift = cv2.xfeatures2d.SIFT_create()
kp1, des1 = sift.detectAndCompute(img1, None)
kp2, des2 = sift.detectAndCompute(img2, None)
```

```python
bf = cv2.BFMatcher()
matcher = bf.match(des1, des2)
matcher = sorted(matcher, key=lambda x : x.distance)
```

```python
img3 = cv2.drawMatches(img1, kp1, img2, kp2, matcher[:100], None, flags=2)
```

```python
my_show(img3)
```

```python
image1_kp = np.float32([kp1[i.queryIdx].pt for i in matcher])
image2_kp = np.float32([kp2[i.trainIdx].pt for i in matcher])
```

```python
H, mask = cv2.findHomography(image2_kp, image1_kp, cv2.RANSAC, 5.0)
```

```python
row,col = np.where(mask == 1)
```
